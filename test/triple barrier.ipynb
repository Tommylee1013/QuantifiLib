```python
from quantifilib.data import YahooFinanceLoader

loader = YahooFinanceLoader()
data = loader.load(
    'SPY',
    start = '2000-01-01',
    interval = '1d',
    auto_adjust = False,
    multi_level_index = False,
    progress = False
)
```

```python
data
```

```python
from quantifilib.strategy.event_driven.triple_barrier import TripleBarrierLabeling

triple_barrier = TripleBarrierLabeling(
    data = data,
    pt_sl = (1, 1),
    min_ret = 0.01,
    vol_lookback = 20,
    vertical_barrier = {
        'days' : 7
    }
)
```

```python
res = triple_barrier.get_labels()
```

```python
res
```

```python
meta_label = TripleBarrierLabeling(
    data = data,
    pt_sl = (1, 1),
    min_ret = 0.01,
    vol_lookback = 20,
    vertical_barrier = {
        'days' : 7
    },
    side_prediction = res['bin']
)
```

```python
meta_label.get_labels()
```
